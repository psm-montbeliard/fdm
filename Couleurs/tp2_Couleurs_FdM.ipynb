## TP2 Codage RVBA

### Slider Alpha

L’objectif est de créer un programme qui permet de modifier la transparence (couche Alpha) d’une image. Vous utiliserez l’une des images proposées par la bibliothèque Skimage. 

La liste est accessible ici : https://scikit-image.org/docs/dev/api/skimage.data.html.

image.png

Une fois l’image choisie, vous pouvez l’importer de la manière suivante : 

```from skimage.data import image_choisie ```

```imageRVB = image_choisie()```


**Attention : assurez-vous que les couleurs de votre image soient codées en RVB.**

Afin de modifier la transparence de l’image dans le programme final, vous devez convertir l’image choisie du mode RVB vers RVBA ; ce qui signifie modifier les dimensions du tableau. Il y a plusieurs solutions possibles, dont celle-ci : 
* Créer un tableau Numpy aux bonnes dimensions pour RVBA ; 
* Copiez dans le tableau Numpy les couches R, V et B de l’image choisie ;
* Le tableau Numpy devient votre image en mode RVBA. 

Enfin, vous afficherez l’image RVBA dans un ImageViewer et ajouterez un slider pour modifier la composante Alpha.
