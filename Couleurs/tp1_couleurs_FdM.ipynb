## Exercice Palette RVB

L’objectif est de créer un ImageViewer contenant une image mélangeant les teintes de rouges et de verts (voir la capture). 

Sans titre2.png

Pour cela, vous créerez une image de taille 256x256, codée en RVB et de couleur entièrement noire. 

Ensuite, vous modifierez chaque pixel de votre image de sorte que : 
* Pour chaque colonne, de gauche à droite, la valeur de Rouge doit aller de 0 à 255 ; 
* Pour chaque ligne, de gauche à droite, la valeur de Vert doit aller de 0 à 255. 

Vous déclinerez votre programme pour les mélanges Rouge-Bleu et Vert-Bleu.

Soit le code suivant pour ajouter un widget (une liste déroulante) à un ImageViewer :

```
from skimage.viewer import ImageViewer 
from skimage.viewer.plugins import Plugin 
from skimage.viewer import widgets 
import numpy as np 

# Créer le viewer et associer une image 
image = np.zeros((10,10,3), dtype=np.float) 
viewer  = ImageViewer(image) 

# Créer le plugin qui va contenir nos widgets 
def changeCouleur(image, choixListe): 
   if choixListe == "coucou": print("coucou") 
   return image 

plugin = Plugin(image_filter = changeCouleur) 

# Ajouter auplugin un widget de type Liste déroulante (ComboBox) 
plugin += widgets.ComboBox("choixListe", ["coucou", "hello"]) 

# Ajouter le plugin au viewer 
viewer += plugin 

# Afficher le viewer 
viewer.show()
```

Vous modifierez le code ci-dessus pour que la liste déroulante permette de choisir entre les modes Rouge-Vert, Rouge-Bleu, Bleu-Vert.

Pour cela, vous créerez d’abord une image noire avec Numpy, codée en RVB. Vous afficherez l’image dans un ImageViewer et ajouterez les trois sliders correspondant aux composantes Rouge, Vert et Bleu.
