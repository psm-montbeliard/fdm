```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2, 100)
plt.plot(x, x, label='linéaire')
plt.plot(x, x**2, label='quadratique')
plt.plot(x, x**3, label='cubique')

plt.xlabel('Etiquette axe OX')
plt.ylabel('Etiquette axe OY')

plt.title("Tracés simples")
plt.legend()
plt.show()
```

```python
p = np.linspace(0.001, .999, 100)
q = 1 - p
```

```python
# h = -p*np.log2(p) - q*np.log2(q)
# entropie : deux états avec les probabilités p et q, p + q = 1
# livre MM, p. 36
plt.axis([0, 1, 0, 1])
plt.plot(p, -p*np.log2(p) - q*np.log2(q),'r-')
plt.show()
```

```python
from numpy import pi
x = np.linspace(0, 2*pi, 100)
y = np.sin(x)
```

```python
#for style in styles:
styles = plt.style.available
print(styles)
style = styles[14]
plt.style.use(style)
plt.axis([0, 2*pi, -1, 1])
plt.plot(x, y)
plt.show()
```

```python
import matplotlib as mpl
angleRad = np.linspace(0, 2*pi, 128)

def demo(quelStyle):
    mpl.style.use(quelStyle)
    fig, ax = plt.subplots(figsize=(3, 3))

    ax.set_title('style: {!r}'.format(quelStyle), color='C0')

    ax.plot(angleRad, np.cos(angleRad), 'C1', label='cos')
    ax.plot(angleRad, np.sin(angleRad), 'C2', label='sin')
    ax.legend()

demo('default')
demo('seaborn')
```
